```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Display the first few rows of the dataset
print(iris.head())

# Summary statistics of the dataset
print(iris.describe())

# Visualization: 
# Pairplot to show the pairwise relationships in the dataset
sns.pairplot(iris, hue='species')
plt.show()

# Boxplot to show the distribution of features for each species
plt.figure(figsize=(10, 5))
sns.boxplot(x='species', y='sepal_length', data=iris)
plt.title('Boxplot of Sepal Length vs Species')
plt.show()

plt.figure(figsize=(10, 5))
sns.boxplot(x='species', y='sepal_width', data=iris)
plt.title('Boxplot of Sepal Width vs Species')
plt.show()

plt.figure(figsize=(10, 5))
sns.boxplot(x='species', y='petal_length', data=iris)
plt.title('Boxplot of Petal Length vs Species')
plt.show()

plt.figure(figsize=(10, 5))
sns.boxplot(x='species', y='petal_width', data=iris)
plt.title('Boxplot of Petal Width vs Species')
plt.show()

# Heatmap to show the correlation between different features
plt.figure(figsize=(8, 8))
sns.heatmap(iris.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Iris Features')
plt.show()
```
